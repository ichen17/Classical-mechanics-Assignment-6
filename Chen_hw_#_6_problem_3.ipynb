1.In the principal axes, the $\Omega$ and $\omega$ can be expressed as

$\overrightarrow{\Omega}=\Omega\sin\left(\alpha\right)\hat{e_{x'}}+\Omega\cos\left(\alpha\right)\hat{e_{z'}}$

$\overrightarrow{\omega}=\omega\cos\left(\alpha\right)\hat{e_{x'}}-\omega\sin\left(\alpha\right)\hat{e_{z'}}=-\Omega\cot\left(\alpha\right)\cos\left(\alpha\right)\hat{e_{x'}}+\Omega\cot\left(\alpha\right)\sin\left(\alpha\right)\hat{e_{z'}}
$

2.The moment of inertia tensor are 

$I=\left(\begin{array}{ccc}
\frac{3}{10}mL^{2}\tan^{2}\alpha & 0 & 0\\
0 & \left(\frac{3}{5}+\frac{3\tan^{2}\alpha}{20}\right)mL^{2} & 0\\
0 & 0 & \left(\frac{3}{5}+\frac{3\tan^{2}\alpha}{20}\right)mL^{2}
\end{array}\right)$

$L=I\omega$

$L=\left(\begin{array}{ccc}
\frac{3}{10}mL^{2}\tan^{2}\alpha & 0 & 0\\
0 & \left(\frac{3}{5}+\frac{3\tan^{2}\alpha}{20}\right)mL^{2} & 0\\
0 & 0 & \left(\frac{3}{5}+\frac{3\tan^{2}\alpha}{20}\right)mL^{2}
\end{array}\right)\left(\begin{array}{c}
\Omega\sin\left(\alpha\right)-\Omega\cot\left(\alpha\right)\cos\left(\alpha\right)\\
0\\
\Omega\cos\left(\alpha\right)+\Omega\cot\left(\alpha\right)\sin\left(\alpha\right)
\end{array}\right)$

$L=\left(\begin{array}{c}
\frac{3}{10}mL^{2}\tan^{2}\alpha\left[\Omega\sin\left(\alpha\right)-\Omega\cot\left(\alpha\right)\cos\left(\alpha\right)\right]\\
0\\
\left(\frac{3}{5}+\frac{3\tan^{2}\alpha}{20}\right)mL^{2}\left[\Omega\cos\left(\alpha\right)+\Omega\cot\left(\alpha\right)\sin\left(\alpha\right)\right]
\end{array}\right)$

3.The formula above is on the principal frame. 

$\overrightarrow{\omega}=-\Omega\cot\left(\alpha\right)\hat{e}_{xb}=-\Omega\cot\left(\alpha\right)\left[\cos\left(\Omega t\right)\hat{e}_{xs}+\sin\left(\Omega t\right)\hat{e}_{ys}\right]$

$\frac{d\overrightarrow{\omega}}{dt}=\Omega^{2}\cot\left(\alpha\right)\left[\sin\left(\Omega t\right)\hat{e}_{xs}+\cos\left(\Omega t\right)\hat{e}_{ys}\right]$

$\frac{d\overrightarrow{L}}{dt}=L_{x'}\frac{\hat{de_{x'}}}{dt}+L_{z'}\frac{\hat{de_{z'}}}{dt}$

becacuse 

$\left(\begin{array}{c}
\hat{e_{x'}}\\
\hat{e_{z'}}
\end{array}\right)=\left(\begin{array}{cc}
\cos\left(\alpha\right) & \sin\left(\alpha\right)\\
-\sin\left(\alpha\right) & \cos\left(\alpha\right)
\end{array}\right)\left(\begin{array}{c}
\hat{e}_{xb}\\
\hat{e}_{zb}
\end{array}\right)$

and $\frac{d\hat{e}_{zb}}{dt}=0$

$\frac{d\overrightarrow{L}}{dt}=\left[\cos\left(\alpha\right)L_{x'}-\sin\left(\alpha\right)L_{z'}\right]\frac{d\hat{e}_{xb}}{dt}$

and since$\frac{d\hat{e}_{xb}}{dt}=\Omega\left[-\sin\left(\Omega t\right)\hat{e}_{xs}+\cos\left(\Omega t\right)\hat{e}_{ys}\right]$

$\frac{d\overrightarrow{L}}{dt}=-\Omega\sin\left(\Omega t\right)\left[\cos\left(\alpha\right)L_{x'}-\sin\left(\alpha\right)L_{z'}\right]\hat{e}_{xs}+\Omega\cos\left(\Omega t\right)\left[\cos\left(\alpha\right)L_{x'}-\sin\left(\alpha\right)L_{z'}\right]\hat{e}_{ys}$

4.Transform it into the body frame,  

$\left(\frac{d\overrightarrow{L}}{dt}\right)_{b}=\Omega\left[\cos\left(\alpha\right)L_{x'}-\sin\left(\alpha\right)L_{z'}\right]\hat{e}_{yb}$

The change of Angular momentum on the body frame is time independent.

5.The sum of change of angular momentum and the torque to cancel the torque created by the gavity is the torque plane applies on the cone 

$N=\frac{d\overrightarrow{L}}{dt}=-\Omega\sin\left(\Omega t\right)\left[\cos\left(\alpha\right)L_{x'}-\sin\left(\alpha\right)L_{z'}-\frac{3}{4}\frac{mgl}{\Omega}\cos\left(\alpha\right)\right]\hat{e}_{xs}+\Omega\cos\left(\Omega t\right)\left[\cos\left(\alpha\right)L_{x'}-\sin\left(\alpha\right)L_{z'}-\frac{3}{4}\frac{mgl}{\Omega}\cos\left(\alpha\right)\right]\hat{e}_{ys}$

```python
import numpy as np
#from scipy.integrate import odeint
import matplotlib.pyplot as plt
#from mpl_toolkits.mplot3d import Axes3D

def angularm(a,m,o,l): #the angular momentum in principal axis
    lll=np.zeros(3)
    ll=m*l*l
    lll[0]=0.3*ll*(np.tan(a)**2)*(o*np.sin(a)-o*np.cos(a)/np.tan(a))
    lll[1]=0
    lll[2]=(0.6+0.15*np.tan(a)**2)*ll*(o*np.cos(a)-o*np.sin(a)/np.tan(a))
    return lll

def N(a,m,o,l):
    g=9.8
    #n=np.zeros(2)
    return o*(np.cos(a)*angularm(a,m,o,l)[0]-np.sin(a)*angularm(a,m,o,l)[2]-m*g*l)-m*g*l*0.75*np.cos(a)

def Na(a,m,o,l):
    g=9.8
    #n=np.zeros(2)
    return o*(np.cos(a)*angularm(a,m,o,l)[0]-np.sin(a)*angularm(a,m,o,l)[2]-m*g*l)#-m*g*l*0.75*np.cos(a)


a=np.pi/6
m=1
l=1
g=9.8
oo=np.linspace(0,500,5000)
NN=np.zeros(5000)
NNa=np.zeros(5000)
for i in range(5000):
    o=i*np.pi/400
    NN[i]=np.abs(N(a,m,o,l))
    NNa[i]=np.abs(Na(a,m,o,l))

plt.title('The magnetitude of torque', fontsize=16)    
plt.plot(oo,NN/g/m/l,label='the torque value')
plt.plot(oo,NNa/g/m/l,label='the approximation')
plt.xlabel('$\Omega$',fontsize=10)
plt.ylabel('$MLg$',fontsize=10)
plt.legend()
plt.show()

plt.title('The relative error', fontsize=16)    
plt.plot(oo,(NN-NNa)/NN)

plt.xlabel('$\Omega$',fontsize=10)
plt.ylabel('$MLg$',fontsize=10)
plt.show()
    
```

The torque function will be like quadratic function. When $\Omega$ is large, the torque from gravity can be ignored.
