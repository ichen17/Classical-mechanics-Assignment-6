For infinitesimal rotation, the rotation matrix can be written down as

$A=\left(1+\epsilon\right)$

$A^{-1}=\left(1-\epsilon\right)$

which $\epsilon$ can written in terms of the generator

$\epsilon=T_{i}d\theta_{i}$

Therefore,

$\frac{dA^{-1}}{dt}A=-\frac{d\epsilon}{dt}\left(1+\epsilon\right)\approx-\frac{d\epsilon}{dt}=-\frac{T_{i}d\theta_{i}}{dt}=-T_{i}\varpi_{i}$

$A=B\left(\psi\right)C\left(\theta\right)D\left(\phi\right)$

$A^{-1}=D^{-1}\left(\phi\right)C^{-1}\left(\theta\right)B^{-1}\left(\psi\right)$

Now, I know

$\frac{dA^{-1}}{dt}=\frac{dD^{-1}\left(\phi\right)}{dt}C^{-1}\left(\theta\right)B^{-1}\left(\psi\right)+D^{-1}\left(\phi\right)\frac{dC^{-1}\left(\theta\right)}{dt}B^{-1}\left(\psi\right)+D^{-1}\left(\phi\right)C^{-1}\left(\theta\right)\frac{dB^{-1}\left(\psi\right)}{dt}$

$\frac{dA^{-1}}{dt}A=-T_{i}\omega_{i}=\frac{dD^{-1}\left(\phi\right)}{dt}D\left(\phi\right)+D^{-1}\left(\phi\right)\frac{dC^{-1}\left(\theta\right)}{dt}C\left(\theta\right)D\left(\phi\right)+D^{-1}\left(\phi\right)C^{-1}\left(\theta\right)\frac{dB^{-1}\left(\psi\right)}{dt}B\left(\psi\right)C\left(\theta\right)D\left(\phi\right)$

Also, we know

$\frac{dD^{-1}\left(\phi\right)}{dt}D\left(\phi\right)=\left(\begin{array}{ccc}
0 & -\dot{\phi} & 0\\
\dot{\phi} & 0 & 0\\
0 & 0 & 0
\end{array}\right)$

$\frac{dC^{-1}\left(\theta\right)}{dt}C\left(\theta\right)=\left(\begin{array}{ccc}
0 & 0 & 0\\
0 & 0 & -\dot{\theta}\\
0 & \dot{\theta} & 0
\end{array}\right)$

$\frac{dB^{-1}\left(\psi\right)}{dt}B\left(\psi\right)=\left(\begin{array}{ccc}
0 & -\dot{\psi} & 0\\
\dot{\psi} & 0 & 0\\
0 & 0 & 0
\end{array}\right)$

$D\left(\phi\right)=\left(\begin{array}{ccc}
\cos\phi & \sin\phi & 0\\
-\sin\phi & \cos\phi & 0\\
0 & 0 & 1
\end{array}\right)$

$C\left(\theta\right)=\left(\begin{array}{ccc}
1 &  0 &  0\\
 0 &  \cos\theta &  \sin\theta\\
 0  &  -\sin\theta &  \cos\theta 
\end{array}\right)$

Therefore, I got 

$-T_{i}\omega_{i}=\left(\begin{array}{ccc}
0 & -\omega_{z} & \omega_{y}\\
\omega_{z} & 0 & -\omega_{x}\\
-\omega_{y} & \omega_{x} & 0
\end{array}\right)=\left(\begin{array}{ccc}
0 & \dot{\phi}+\dot{\psi}\cos\left(\theta\right) & +\dot{\psi}\sin\theta\cos\phi-\sin\phi\dot{\theta}\\
-\dot{\phi}-\dot{\psi}\cos\left(\theta\right) & 0 & \cos\phi\dot{\theta}+\dot{\psi}\sin\theta\sin\phi\\
-\dot{\psi}\sin\theta\cos\phi+\sin\phi\dot{\theta} & -\cos\phi\dot{\theta}-\dot{\psi}\sin\theta\sin\phi & 0
\end{array}\right)$

Therefore, we can know

$\omega_{x}=\cos\phi\dot{\theta}+\dot{\psi}\sin\theta\sin\phi$

$\omega_{y}=-\dot{\psi}\sin\theta\cos\phi+\sin\phi\dot{\theta}$

$\omega_{z}=\dot{\phi}+\dot{\psi}\cos\left(\theta\right)$

In space frame, $\omega_{\phi}$ is parallel to space axis. In body frame, it becomes

$\left(\begin{array}{ccc}
\cos\psi & \sin\psi & 0\\
-\sin\psi & \cos\psi & 0\\
0 & 0 & 1
\end{array}\right)\left(\begin{array}{ccc}
1 & 0 & 0\\
0 & \cos\theta & \sin\theta\\
0 & -\sin\theta & \cos\theta
\end{array}\right)\left(\begin{array}{c}
0\\
0\\
\dot{\phi}
\end{array}\right)=\left(\begin{array}{c}
\dot{\phi}\sin\psi\sin\theta\\
\dot{\phi}\cos\psi\sin\theta\\
\dot{\phi}\cos\theta
\end{array}\right)$

for $\omega_{\theta}$, in body frame, it becomes

$\left(\begin{array}{ccc}
\cos\psi & \sin\psi & 0\\
-\sin\psi & \cos\psi & 0\\
0 & 0 & 1
\end{array}\right)\left(\begin{array}{c}
\dot{\theta}\\
0\\
0
\end{array}\right)=\left(\begin{array}{c}
\cos\psi\dot{\theta}\\
-\sin\psi\dot{\theta}\\
0
\end{array}\right)$

For $\omega_{\psi}$, 

$\left(\begin{array}{c}
0\\
0\\
\dot{\psi}
\end{array}\right)$

Therefore,

$\omega_{x}	=\dot{\phi}\sin\psi\sin\theta+\cos\psi\dot{\theta}$

$\omega_{y}	=\dot{\phi}\cos\psi\sin\theta-\sin\psi\dot{\theta}$

$\omega_{z}	=\dot{\phi}\cos\theta+\dot{\psi}$

```python
import numpy as np
#from scipy.integrate import odeint
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def spframe(psi,theta,phi,dpsi,dtheta,dphi): # anuglar momentum in space frame.
    o=np.ones(3)
    o[0]=dtheta*np.cos(phi)+dpsi*np.sin(theta)*np.sin(phi)
    o[1]=-dpsi*np.sin(theta)*np.cos(phi)+dtheta*np.sin(phi)
    o[2]=dphi+dpsi*np.cos(theta)
    return o
    
def bframe(psi,theta,phi,dpsi,dtheta,dphi): # anuglar momentum in body frame.
    o=np.ones(3)
    #o1=np.ones(3)
    o[0]=dtheta*np.cos(psi)+dphi*np.sin(theta)*np.sin(psi)
    o[1]=dphi*np.sin(theta)*np.cos(psi)-dtheta*np.sin(psi)
    o[2]=dpsi+dphi*np.cos(theta)
    #cps=np.cos(-phi)
    #sps=np.sin(-phi)
    #cph=np.cos(-psi)
    #sph=np.sin(-psi)
    #ct=np.cos(-theta)
    #st=np.sin(-theta)
    #o1[0]=(cps*cph-ct*sps*sph)*o[0]+(cps*sph+ct*cph*sps)*o[1]+(sps*st)*o[2]
    #o1[1]=(-sps*cph-ct*cps*sph)*o[0]+(-sps*sph+ct*cph*cps)*o[1]+(cps*st)*o[2]
    #o1[2]=(st*sph)*o[0]+(-st*cph)*o[1]+(ct)*o[2]
    return o

def sp_bframe(psi,theta,phi,dpsi,dtheta,dphi): #transform angular momentum in space frame to body frame
    o=np.ones(3)
    o1=np.ones(3)
    o[0]=dtheta*np.cos(phi)+dpsi*np.sin(theta)*np.sin(phi)
    o[1]=-dpsi*np.sin(theta)*np.cos(phi)+dtheta*np.sin(phi)
    o[2]=dphi+dpsi*np.cos(theta)
    cps=np.cos(psi)
    sps=np.sin(psi)
    cph=np.cos(phi)
    sph=np.sin(phi)
    ct=np.cos(theta)
    st=np.sin(theta)
    o1[0]=(cps*cph-ct*sps*sph)*o[0]+(cps*sph+ct*cph*sps)*o[1]+(sps*st)*o[2]
    o1[1]=(-sps*cph-ct*cps*sph)*o[0]+(-sps*sph+ct*cph*cps)*o[1]+(cps*st)*o[2]
    o1[2]=(st*sph)*o[0]+(-st*cph)*o[1]+(ct)*o[2]
    return o1

    
```
